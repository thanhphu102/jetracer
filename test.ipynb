```python
import time
from mpu9250_jmdev.registers import *
from mpu9250_jmdev.mpu_9250 import MPU9250

# Khởi tạo đối tượng MPU9250
# Đảm bảo tham số 'bus' khớp với bus I2C bạn đã quét ở Bước 1 (thường là 1)
mpu = MPU9250(
    address_ak=AK8963_ADDRESS, 
    address_mpu_master=MPU9050_ADDRESS_68, # Thay bằng 69 nếu i2cdetect ra 69
    address_mpu_slave=None, 
    bus=1, 
    gfs=GFS_1000, 
    afs=AFS_8G, 
    mfs=AK8963_BIT_16, 
    mode=AK8963_MODE_C100HZ)

# Cấu hình cảm biến trước khi đọc
mpu.configure()

print("Bắt đầu đọc dữ liệu từ MPU9250. Nhấn Ctrl+C để thoát.\n")

try:
    while True:
        # Đọc dữ liệu Gia tốc (Accelerometer) - Đơn vị: g
        accel = mpu.readAccelerometerMaster()
        
        # Đọc dữ liệu Con quay hồi chuyển (Gyroscope) - Đơn vị: deg/s
        gyro = mpu.readGyroscopeMaster()
        
        # Đọc dữ liệu Từ kế (Magnetometer) - Đơn vị: uT (microTesla)
        mag = mpu.readMagnetometerMaster()
        
        # Đọc Nhiệt độ môi trường cảm biến - Đơn vị: Độ C
        temp = mpu.readTemperatureMaster()

        # In kết quả
        print(f"Gia tốc (X, Y, Z):      {accel[0]:.2f}, {accel[1]:.2f}, {accel[2]:.2f} (g)")
        print(f"Con quay (X, Y, Z):     {gyro[0]:.2f}, {gyro[1]:.2f}, {gyro[2]:.2f} (°/s)")
        print(f"Từ trường (X, Y, Z):    {mag[0]:.2f}, {mag[1]:.2f}, {mag[2]:.2f} (µT)")
        print(f"Nhiệt độ:               {temp:.2f} °C")
        print("-" * 40)
        
        time.sleep(0.5)

except KeyboardInterrupt:
    print("\nĐã dừng đọc dữ liệu.")
```
